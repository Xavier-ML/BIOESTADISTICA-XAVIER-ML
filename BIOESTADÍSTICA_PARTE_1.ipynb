##Prevenci√≥n de endometriosis
La endometriosis es una afecci√≥n a menudo dolorosa en la cual un tejido similar al que recubre el interior del √∫tero crece fuera de este. Con frecuencia afecta a los ovarios, las trompas de Falopio y el tejido que recubre la pelvis. Es por ello que el Municipio realizar√° campa√±as de prevenci√≥n si se encuentra que la edad promedio en que la que mujeres son diagnosticadas con dicha enfermedad es menor a 24 a√±os. ¬øCu√°l ser√° la decisi√≥n del Municipio considerando los siguientes datos? Use un nivel de significaci√≥n del 5%.
22,24,21,20,26,28,22,21,18,13,23,27,29,16,31,19

### Hip√≥tesis
Ho: u >= 24 (Hip√≥tesis nula)

Ha: u < 24 (Hip√≥tesis alterna)

alfa = 0.05

```r
datos<-c(22,24,21,20,26,28,22,21,18,13,23,27,29,16,31,19)
summary(datos) # resumen estaditico de la muestra
hist(datos) # grafica de frecuencias en barra conocido como histograma
x_bar<-mean(datos) # calculo de ma media de los datos
s<-sd(datos) # calculo de la desviaci√≥n estandar "s" de los datos
n<-length(datos) # calculo de la cantidad de datos de la muestra
# calculo de estadistico de prueba
tcal<- (x_bar-24)/(s/sqrt(n)) # Este es el t que mide la variabilidad de la media, la formula esta en la parte de arriba
tcal
# valor critico
qt(0.05,n-1) #‚Äòqt‚Äô gives the quantile function, representa el valor de p teorico poblacional, en este cao nos permite calcular el valor critico referencial.
```

### P-Valor

```r
# calculo de pvalor por diferentes maneras
pt(tcal,n-1) # ‚Äòpt‚Äô gives the distribution function, representa el valor de p valor calculado de la muestra
t.test(datos,mu=24,alternative="l")

# Verificar supuesto de Normalidad
#Ho: los datos se ajustan a una distribuci√≥n normal
#H1: Los datos No se ajustan a una distribuci√≥n normal
# los siguientes test tienen sus propios argmentos de uso que se tocar√° m√°s a delante.
# pruebas de normalidad sobre los datos
shapiro.test(datos)
install.packages("nortest")
library(nortest)
ad.test(datos)
cvm.test(datos)
set.seed(2024)

install.packages("BSDA")
library(BSDA)
SIGN.test(datos, md =24,alternative = "less",
          conf.level = 0.95)

install.packages("exactRankTests")
library(exactRankTests)
wilcox.exact(datos,mu=24,alternative="less")
# en todos los casos no se rechaza la hip√≥tesis nula
```
