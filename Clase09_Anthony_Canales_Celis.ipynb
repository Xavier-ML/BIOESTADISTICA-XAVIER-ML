```python
#1. Describir variables cuantitativas
import pandas as pd

x = pd.read_csv("/content/train.csv")

pd.DataFrame(x,columns=["Survived","PClass","Age","Sibsp","Parch","Fare"])
```

```python
#2. Graficar supervivencia por género
var1=x.groupby("Sex")["Survived"].sum()
var1.plot.bar()
```

```python
#3. Graficar supervivencia por edad
var2=x.groupby("Age")["Survived"].sum()
var2.plot.bar()
```

```python
#4. encontrar la relación entre el precio y la supervivencia
x["relacion_Precio_Sobrevivencia"]=x.Fare/x.Survived
x
```

```python
# 5. encontrar la relación entre el precio y la clase
var4=x.groupby("Survived")["Fare"].describe()
var2.plot.barh()
```

```python
#6. Extraer los títulos de los nombres(Name), ejemplo: Mr

def split_it(Name):
  var = Name.split()
  title= ""
  if "Mr." in var:
    title = "Mr."
  elif "Miss." in var:
    title = "Miss."
  elif "Mrs." in var:
    title = "Mrs."
  else:
    title = "No definido"
  return title

x["Title"]=x.Name.apply(lambda x : split_it(x))
x
```
